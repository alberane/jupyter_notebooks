# Custerização usando KMeans

Queremos uma maneira de agrupar as amostras automaticamente, sem a necessidade de "treinar" o algoritmo. ou seja, estamos procurando um algoritmo de **aprendizado não supervisionado**.

O **KMeans Clustering** é um desses algo de ***Aprendizagem Não Supervisionada,*** que, olhando os dados, agrupa as amostras em '**clusters**' com base na distância de cada amostra do centro do grupo.

Aqui pode ser uma boa fonte de consulta.
[scikit-learn.org - KMeans](https://scikit-learn.org/stable/modules/clustering.html#k-means)

```
%matplotlib inline

import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
```

```
iris = load_iris()
#iris.data[0:5]
#iris.target
```

```
kmeans = KMeans(n_clusters=3)
```

```
kmeans_model = kmeans.fit(iris.data)
#kmeans_model.labels_
#kmeans_model.cluster_centers_
```

```
pd.crosstab(iris.target, kmeans_model.labels_)
```

```
X = iris.data[:, :2] #usa apenas uma fatia dos dados
y = iris.target
```

```
plt.scatter(X[:,0], X[:,1], c=y, cmap='gist_rainbow')
plt.xlabel('Tamanho Sépala', fontsize=12)
plt.ylabel('Largura Sépala', fontsize=12)
```

```
X2 = iris.data[:, -2:] #usa a outra fatia dos dados
y2 = iris.target
```

```
plt.scatter(X2[:,0], X2[:,1], c=y2, cmap='gist_rainbow')
plt.xlabel('Tamanho Pétala', fontsize=12)
plt.ylabel('Largura Pétala', fontsize=12)
```
